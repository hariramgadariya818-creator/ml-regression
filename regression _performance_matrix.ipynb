```python
import numpy as np
from sklearn.metrics import *
```

```python
y_true = np.array([17, 15, 14, 14, 19])
y_pred = np.array([16, 10, 13, 20, 18])
y_t=sum(y_true)
y_p=sum(y_pred)
```

```python
# MAE
mae = mean_absolute_error(y_true, y_pred)
print(mae)

n=len(y_true)
mae_v= np.sum(np.abs(y_true - y_pred)) / n
print(mae_v)

# Mean Absolute Error (MAE) measures the average absolute difference between the actual values and predicted values. It tells us how much the model’s predictions deviate from the true values on average.
```

```python
# MSE
mse = mean_squared_error(y_true, y_pred)
print(mse)

mse_v=np.sum((y_true - y_pred)**2)/n
print(mse_v)

# Mean Squared Error (MSE) calculates the average of the squared differences between actual and predicted values. Because the errors are squared, larger errors have a much greater impact than smaller ones.
```

```python
# RMSE
rmse = np.sqrt(mean_squared_error(y_true, y_pred))
print(rmse)

rmse_v = np.sqrt(np.sum((y_true - y_pred)**2)/n)
print(rmse_v)

# Root Mean Squared Error (RMSE) is the square root of MSE. It converts the squared error back into the original unit of the data, making it easier to interpret.
```

```python
# MAPE
mape = mean_absolute_percentage_error(y_true, y_pred)
print(mape)

mape_v = np.mean(np.abs((y_true - y_pred) / y_true)) *100
print(mape_v)

# Mean Absolute Percentage Error (MAPE) measures the average percentage difference between actual and predicted values.
```

```python
# R2 Score
r2 = r2_score(y_true, y_pred)
print(r2)

# mean of actual values
y_mean = np.mean(y_true)
# SSE (sum of squared errors)
sse = np.sum((y_true - y_pred)**2)
# SST (total variance)
sst = np.sum((y_true - y_mean)**2)
# calculate R2
r2_v = 1 - (sse / sst)

print(r2_v)

# model 76% data samajh gaya
# R² Score measures how well the regression model explains the variance in the data. It indicates how much of the variation in the dependent variable is explained by the model.
```

```python
# Adjusted R2
p = 1               # kitne features h 
adjusted_r2 = 1 - (1-r2)*(n-1)/(n-p-1)
print(adjusted_r2)

# Adjusted R² improves the R² score by considering the number of features used in the model. It penalizes unnecessary features, giving a more realistic measure of model performance.
```
